```python
import numpy as np
import pandas as pd
count =0
df = pd.DataFrame(columns=['A','B','C','F(A)','F(B)','F(B)','Error'])

def bisection(num1,num2):
    return (num1+num2)/2

def equation(x):
    return (x**2)-3

Tresh = float(input("Enter Error: "))
while True:
    a = float(input("Enter A Value: "))
    b = float(input("Enter B Value: "))
    ffa = equation(a)
    ffb = equation(b)
    if ((ffa*ffb)>0):
        print("Intermediate Theoram could apply because one of the function of value should return positive and other negative")
    else:
        break

while True:
    c = bisection(a,b)
    df.loc[count] = [a,b,c,equation(a),equation(b),equation(c),(c-a)]
    if (equation(c) * equation(a))>0:
        a = c
    else:
        b = c
    count = count+1
    if np.abs(float(a-b)) <= Tresh:
        break
        
df
```
