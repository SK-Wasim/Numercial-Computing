```python
import numpy as np
import pandas as pd

df = pd.DataFrame(columns=['x1','x2','x3'])
c = np.array([[2,-1,1,-1],[1,2,-1,6],[1,-1,2,-3]])
x = np.ones(3)

x = np.zeros(len(c))
xcop = np.zeros(len(c))
for k in range(0,15):
    for i in range(0,len(c)):
        y = 0
        for j in range(0,len(c)):
            if i != j:
                y = y + (c[i][j] * x[j])
        y = c[i][len(c)] - y
        xcop[i] = y / c[i][i]
    for i in range(0,len(x)):
        x[i] = xcop[i]
    df.loc[k+1] = [x[0],x[1],x[2]]
        
round(df,4)
```
