```python
xValues = df.drop('Y house price of unit area', axis=1).values
yValues = df['Y house price of unit area'].values
```

```python
length = len(xValues)//5
x_train,x_test = np.split(xValues,[(length*4)],axis=0)
y_train,y_test = np.split(yValues,[(length*4)],axis=0)
```

```python
def gradientDescent(x,y,_i,_l):
    m1 = 0
    m2 = 0
    m3 = 0
    m4 = 0
    c = 1
    l = _l
    n = len(x)
    for i in range(_i):
        y_pred = m1*x[:,0] + m2*x[:,1] + m3*x[:,2] + m4*x[:,3] + c
        cost = (1/(2*n))*sum([val**2 for val in (y-y_pred)])
        D_m1 = (-2/n) * sum(x[:,0] * (y - y_pred))
        D_m2 = (-2/n) * sum(x[:,1] * (y - y_pred))
        D_m3 = (-2/n) * sum(x[:,2] * (y - y_pred))
        D_m4 = (-2/n) * sum(x[:,3] * (y - y_pred))
        D_c = (-2/n) * sum(y - y_pred)
        m1 = m1 - l * D_m1
        m2 = m2 - l * D_m2
        m3 = m3 - l * D_m3
        m4 = m4 - l * D_m4
        c = c - l * D_c
    return m1,m2,m3,m4,c,cost
```
