```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(columns=['x1','x2','x3','x4'])
c = np.array([[10,-1,2,0,6],[-1,11,-1,3,25],[2,-1,10,-1,-11],[0,3,-1,8,15]])

x = np.zeros(len(c))
for k in range(0,9):
    for i in range(0,len(c)):
        y = 0
        for j in range(0,len(c)):
            if i != j:
                y = y + (c[i][j] * x[j])
        y = c[i][len(c)] - y
        x[i] = y / c[i][i] 
    df.loc[k+1] = [x[0],x[1],x[2],x[3]]
        
round(df,4)
```
